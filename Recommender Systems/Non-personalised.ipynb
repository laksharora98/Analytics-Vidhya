Q3

```python
data.head()
```

```python
data_article.head()
```

```python
data_rating.columns = ['average_rating']
data_rating.head()
```

```python
data_q3 = data_article.merge(right=data_rating, how='left', on='article_id')
```

```python
data_q3
```

```python
m = 2
data_q3 = data_q3.loc[data_q3.number_users >= m]
```

```python
c = data['rating'].mean()
c
```

```python
data_q3['weighted_rating'] = (data_q3.average_rating*data_q3.number_users+c*m)/(data_q3.number_users + m)
```

```python
data_q3.sort_values(by=['weighted_rating'], ascending=False).head(10)
```
